## We would like to get the population per continent (not country this time).

Get the continent year and population (country here is optional and for testing purposes).

```python
data3 = data[['continent', 'country', 'year', 'pop']]
```

Now, show that countries are unique.  For the year 1952, we have unique countries.  Here, Algeria and 1952 = 1 row.  This is the reason we were able to do a pivot on this dataset.

```python
data3.query('year == 1952 and country=="Algeria"')
```

This shows continents are NOT unique. Pivoting requires the index to be unique.  Here, Africa contains Algeria but also other countries, so it's not unique.  We have to make this list unique somehow by continent.

```python
data3.query('year == 1952 and continent == "Africa"').head(2)
```

```python
data3.head(2)
```

__Here is the fix__:  We group by continent and year, and we include the population sum for the group as an aggregate value using the agg function.

```python
data3_grouped = data3.groupby(['continent', 'year'], as_index=False).agg({'pop': 'sum'})
```

In preparation for showing the results, I foresee the population values to be really huge.  So to avoid seeing them in scientific notation, use the following "set option" feature of Pandas.

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

So what do we get?  Here is a sample of the result.  The results are correct, and it's sorted by continent.   Use _data.sort_values('year', inplace=true)_ if you want to sort otherwise.

But, there is a little issue if we want to be perfectionists.  The index here is not sorted.  Usually it doesn't matter but use a _reset_index_ command to fix this as shown below.

```python
data3_grouped.head(10)
```

```python
data3_grouped.sort_values('year', inplace=True)
```

Note:  When resetting the index, we need to create a new dataset if you want to retain this new sorting.  (Or you can just change the same dataset by using the same name, A = A.reset_index).

```python
data3_grouped_sorted_by_year = data3_grouped.reset_index(drop=True)
```

Here's the final result.

```python
data3_grouped_sorted_by_year.head(10)
```
