```python
import pandas as pd
```

```python
stat1 = pd.read_csv("http://bit.ly/2cLzoxH")
```

```python
df = stat1[['continent','lifeExp']]
```

```python
df.head(3)
```

```python
print(df.shape)
```

```python
pd.pivot_table(df, values='lifeExp', columns='continent')
```

```python
df = stat1[['country','continent','lifeExp','year']]
```

```python
pd.pivot_table(df, values='lifeExp', columns='country')
```

```python
pd.pivot_table(df, values='lifeExp', columns='country')['Albania']
```

```python
pd.pivot_table(stat1, values='lifeExp', index=['year'], columns='continent')
```

```python
pd.pivot_table(stat1, values='lifeExp', index=['year'], columns='continent').index
```

```python
pd.pivot_table(stat1, values='lifeExp', index=['year'], columns='continent')['Americas']
```

```python
pd.pivot_table(stat1, values='lifeExp',
                     index=['year'],
                     columns='continent',
                     aggfunc=[min,max])
```
